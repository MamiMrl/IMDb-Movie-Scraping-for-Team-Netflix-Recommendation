```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd

# load dataframe_movies
dataframe_movies = pd.read_csv("archive/movies_metadata.csv")

# create a new DataFrame to store the extracted information
df_movie_info = pd.DataFrame(columns=["movie_name", "rating", "num_ratings", "budget"])

# start a Chrome webdriver
driver = webdriver.Chrome()

# go to the IMDb homepage
url = "https://www.imdb.com/?ref_=nv_home"
driver.get(url)

# locate the search bar element
search_bar = driver.find_element(By.ID, "suggestion-search")

# iterate through each row of the imdb_id column of dataframe_movies
for imdb_id in dataframe_movies["imdb_id"]:
    # enter the imdb_id value into the search bar
    search_bar.clear()
    search_bar.send_keys(imdb_id)
    search_bar.submit()

    # extract the required information from the search results page
    soup = BeautifulSoup(driver.page_source, "html.parser")
    movie_name = soup.find("span", class_="sc-afe43def-1 fDTGTb").text.strip()
    rating = soup.find("span", class_="sc-bde20123-1 iZlgcd").text.strip()
    num_ratings = soup.find("div", class_="sc-bde20123-3 bjjENQ").text.strip()
    budget = soup.findAll('span', {'class': 'ipc-metadata-list-item__list-content-item'})[2].text.strip().split(' (')[0]

    # add the extracted information to the DataFrame
    df_movie_info = df_movie_info.append({
        "movie_name": movie_name,
        "rating": rating,
        "num_ratings": num_ratings,
        "budget": budget
    }, ignore_index=True)

    # navigate back to the homepage
    driver.back()
    search_bar = driver.find_element(By.ID, "suggestion-search") # locate the search bar element again

# close the Chrome webdriver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd

# load dataframe_movies
dataframe_movies = pd.read_csv("archive/movies_metadata.csv")

# create a new DataFrame to store the extracted information
df_movie_info = pd.DataFrame(columns=["movie_name", "rating", "num_ratings", "budget"])

# start a Chrome webdriver
driver = webdriver.Chrome()

# go to the IMDb homepage
url = "https://www.imdb.com/?ref_=nv_home"
driver.get(url)

# locate the search bar element
search_bar = driver.find_element(By.ID, "suggestion-search")

# iterate through each row of the imdb_id column of dataframe_movies
for imdb_id in dataframe_movies["imdb_id"]:
    # enter the imdb_id value into the search bar
    search_bar.clear()
    search_bar.send_keys(imdb_id)
    search_bar.submit()

    # extract the required information from the search results page
    soup = BeautifulSoup(driver.page_source, "html.parser")
    
    try:
        movie_name = soup.find("span", class_="sc-afe43def-1 fDTGTb").text.strip()
    except IndexError:
        movie_name = None
    
    try:
        rating = soup.find("span", class_="sc-bde20123-1 iZlgcd").text.strip()
    except IndexError:
        rating = None
    
    try:
        num_ratings = soup.find("div", class_="sc-bde20123-3 bjjENQ").text.strip()
    except IndexError:
        num_ratings = None
    
    try:
        budget = soup.findAll('span', {'class': 'ipc-metadata-list-item__list-content-item'})[2].text.strip().split(' (')[0]
    except IndexError:
        budget = None

    # add the extracted information to the DataFrame
    df_movie_info = df_movie_info.append({
        "movie_name": movie_name,
        "rating": rating,
        "num_ratings": num_ratings,
        "budget": budget
    }, ignore_index=True)

    # navigate back to the homepage
    driver.back()
    search_bar = driver.find_element(By.ID, "suggestion-search") # locate the search bar element again

# close the Chrome webdriver
driver.quit()
```
